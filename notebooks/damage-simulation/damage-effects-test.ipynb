```python
# Ensure src is in the import path
import os
import sys

project_root = os.path.abspath("..")
src_path = os.path.join(project_root, "src")
if src_path not in sys.path:
    sys.path.append(src_path)
```

```python
from PIL import Image
import matplotlib.pyplot as plt

# Import damage classes
from damage_effects import (TearDamage, ScratchDamage, 
                            WaterDiscolouration, FlakingDamage,
                            CraquelureDamage)
```

# Load the sample image

```python
# Load the base image
input_path = os.path.join(project_root,
                          "data", 
                          "sample-image-canvas-oil-Le-Moulin-de-la-Galette-1876.png")
original_image = Image.open(input_path).convert("RGB")
```

# Apply the damages

```python
# Define damage versions
damage_versions = [
    [],  # Original
    [TearDamage()],
    [ScratchDamage()],
    [WaterDiscolouration()],
    [FlakingDamage()],
    [CraquelureDamage()],
]

# Generate images
images = []
for pipeline in damage_versions:
    img = original_image.copy()
    for dmg in pipeline:
        img = dmg.apply(img)
    images.append(img)
```

# Visualize the damages

```python
titles = [
    "Original",
    "TearDamage",
    "ScratchDamage",
    "WaterDiscolouration",
    "FlakingDamage",
    "CraquelureDamage"
]

# Plot as grid
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

for i, ax in enumerate(axes.flat):
    ax.imshow(images[i])
    ax.set_title(titles[i])
    ax.axis("off")

plt.tight_layout()
plt.show()
```
